```python
from rdkit import Chem 
df2['mol'] = df2['F[P-](F)(F)(F)(F)F.CCCC[N+]1=CC=CC(C)=C1'].apply(lambda x: Chem.MolFromSmiles(x)) 
print(type(df2['mol'][0]))
```

```python
from rdkit.Chem import Draw
mols = df2['mol'][:20]
Draw.MolsToGridImage(mols, molsPerRow=5, useSVG=True, legends=list(df2['F[P-](F)(F)(F)(F)F.CCCC[N+]1=CC=CC(C)=C1'][:20].values))
```
